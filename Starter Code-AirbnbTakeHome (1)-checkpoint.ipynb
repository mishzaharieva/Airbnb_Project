## Overview

    
Airbnb is a two sided marketplace which matches guests to hosts. The booking flow at Airbnb is as follows: a guest
finds an available room (listing) that they like, and then they contact the host. Once the guest finds a listing they
are interested in, there are three ways to send the host an inquiry: contact_me, book_it, or instant_book
(detailed at the bottom of this document). Upon receiving the inquiry, the host can then decide whether or not to
accept the request (for contact_me and book_it methods -- instant_book is auto-accepted). One of our goals at
Airbnb is to increase bookings on our platform.
    
You are the first data scientist to join a cross-functional Product and Operations team working to grow bookings in
Rio de Janeiro. The team asks you for help with the following:

1. What key metrics would you propose to monitor over time the success of the team's efforts in improving
the guest host matching process and why? Clearly define your metric(s) and explain how each is
computed.

segments are doing well and what could be improved? Propose 2-3 specific recommendations (business
initiatives and product changes) that could address these opportunities. Demonstrate rationale behind
each recommendation AND prioritize your recommendations in order of their estimated impact.
3. There is also interest from executives at Airbnb about the work you are doing, and a desire to understand
the broader framing of the challenge of matching supply and demand, thinking beyond the data provided.
What other research, experiments, or approaches could help the company get more clarity on the
problem?

Your assignment: Summarize your recommendations in response to the questions above in a 5-8 slide
presentation intended for the Head of Product and VP of Operations (who is not technical). Include an organized
appendix sharing the details of your work conducted for the Rio team, that would be useful for the data team to

```python
null_counts = airbnb_df.isnull().sum()
print(null_counts)
```

```python
# Upon investigation, it was found that 219 rows of missing data resulted from merging the listings and contacts data. 
# These missing values were subsequently dropped from the airbnb_df to ensure data integrity for analysis.
# The approach involved iterating through the DataFrame and directly removing these rows to effect changes within the same dataset, avoiding the creation of a new DataFrame with dropped rows.
```

```python
for i in airbnb_df.index:
    if pd.isnull(airbnb_df.loc[i, 'ds_checkin_first']) and \
       pd.isnull(airbnb_df.loc[i, 'ds_checkout_first']) and \
       pd.isnull(airbnb_df.loc[i, 'm_guests']) and \
       pd.isnull(airbnb_df.loc[i, 'm_interactions']) and \
       pd.isnull(airbnb_df.loc[i, 'm_first_message_length_in_characters']) and \
       pd.isnull(airbnb_df.loc[i, 'contact_channel_first']) and \
       pd.isnull(airbnb_df.loc[i, 'guest_user_stage_first']):
        airbnb_df.drop(i, inplace=True)
airbnb_df.reset_index(drop=True, inplace=True)
```

```python
# Certain columns crucial for analysis, such as ts_interaction_first, ts_accepted_at_first, ts_booking_at, and ts_reply_at_first, are not subjected to data imputation.
```

```python
# Creation of Binary Indicator Columns
# booking_completed: Assigns 0 when ts_booking_at is null and 1 when it contains a date value.
# host_replied: Assigns 0 when ts_reply_at_first is null and 1 when it contains a date value.
# host_accepted: Assigns 0 when ts_accepted_at_first is null and 1 when it contains a date value.
# interaction_occurred: Assigns 0 when ts_interaction_first is null and 1 when it contains a date value.
```

```python
airbnb_df['booking_completed'] = airbnb_df['ts_booking_at'].notnull().astype(int)
```

```python
airbnb_df['host_replied'] = airbnb_df['ts_reply_at_first'].notnull().astype(int)
```

```python
airbnb_df['host_accepted'] = airbnb_df['ts_accepted_at_first'].notnull().astype(int)
```

```python
airbnb_df['interaction_occurred'] = airbnb_df['ts_interaction_first'].notnull().astype(int)
```

```python
null_counts = airbnb_df.isnull().sum()
print(null_counts)
```

```python
# The only remaining null value is in the m_guests column. To address this, I will input a 0 to preserve all data for the analysis, ensuring that no rows are deleted.
```

```python
airbnb_df['m_guests'].fillna(0, inplace=True)
```

```python
null_counts = airbnb_df.isnull().sum()
print(null_counts)
```

```python
# Checking the dataframe for duplicate entries.
```

```python
airbnb_df.duplicated().sum()
```

```python
# Creation of Calculated Columns
 # response_time: Compute the difference between ts_reply_at_first and ts_interaction_first. Convert the result to minutes by dividing the total seconds by 60.
 # booking_time: Compute the difference between ts_accepted_at_first and ts_interaction_first. Similarly, convert the result to minutes by dividing the total seconds by 60.
```

```python
airbnb_df['ts_interaction_first'] = airbnb_df['ts_interaction_first'].astype(str)
airbnb_df['ts_interaction_first'] = airbnb_df['ts_interaction_first'].str.replace('.0', '', regex=False)
airbnb_df['ts_interaction_first'] = pd.to_datetime(airbnb_df['ts_interaction_first'])
airbnb_df['ts_reply_at_first'] = pd.to_datetime(airbnb_df['ts_reply_at_first'])
airbnb_df['ts_accepted_at_first'] = pd.to_datetime(airbnb_df['ts_accepted_at_first'])
airbnb_df['response_time'] = (airbnb_df['ts_reply_at_first'] - airbnb_df['ts_interaction_first']).dt.total_seconds() / 60
airbnb_df['booking_time'] = (airbnb_df['ts_accepted_at_first'] - airbnb_df['ts_interaction_first']).dt.total_seconds() / 60

print(airbnb_df.head())
```

```python
null_counts = airbnb_df.isnull().sum()
print(null_counts)
```

```python
airbnb_df['response_time'].fillna(0, inplace=True)
airbnb_df['booking_time'].fillna(0, inplace=True)
```

```python
# Hyphothesis 1: I will calculate the total bookings and the completed bookings for each contact channel. 
# Then, I will plot these values onto a bar chart to analyze the effectiveness of different contact channels.
```

```python
total_bookings = airbnb_df.groupby('contact_channel_first')['booking_completed'].count().reset_index()
bookings_completed = airbnb_df.groupby('contact_channel_first')['booking_completed'].sum().reset_index()
```

```python
total_bookings = total_bookings.sort_values(by='booking_completed', ascending=False)
fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.35
r1 = np.arange(len(total_bookings['contact_channel_first']))
r2 = [x + bar_width for x in r1]

bar1 = ax.bar(r1, total_bookings['booking_completed'], color='#FF5A5F', width=bar_width, label='Total Inquiries')
bar2 = ax.bar(r2, bookings_completed['booking_completed'], color='#00A699', width=bar_width, label='Bookings Completed')

for bar in bar1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 200, int(yval), ha='center', va='bottom', fontsize=8)

for bar in bar2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 200, int(yval), ha='center', va='bottom', fontsize=8)

ax.set_xlabel('Contact Channel')
ax.set_ylabel('Number of Bookings')
ax.set_title('Instant Book Channel Leads with Highest Booking Completion Rate Despite Lower Total Inquiries', fontsize=12, pad=18)

fig.suptitle('Comparison of Total Inquiries and Completed Bookings by Contact Channel', fontsize=14)

ax.set_xticks([r + bar_width / 2 for r in range(len(total_bookings['contact_channel_first']))])
ax.set_xticklabels(total_bookings['contact_channel_first'], rotation=45)

ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplots_adjust(top=0.9)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Findings from this bar chart may show that most inquiries come through the Contact Me method, yet this channel results in the fewest successful bookings. 
# This supports the hypothesis that the Contact Me channel is less effective for securing bookings.
```

```python
# Performing Cross Validation
```

```python
# Prepare Data with xgb.DMatrix
```

```python
import xgboost as xgb
dtrain = xgb.DMatrix(data=x_train, label=y_train)
dtest = xgb.DMatrix(data=x_test)
```

```python
# Define Parameters
```

```python
params = {
    "objective": "binary:logistic",
    "max_depth": 4,
    "learning_rate": 0.1,
    "subsample": 0.8,
    "colsample_bytree": 0.8,
    "eval_metric": "error"
}
```

```python
# Perform Cross-Validation
```

```python
cross_val_results = xgb.cv(
    params=params,
    dtrain=dtrain,
    num_boost_round=100,
    nfold=5,
    metrics="error",
    early_stopping_rounds=10,
    seed=42,
    as_pandas=True
)
```

```python
# Evaluate Cross-Validation Results
```

```python
print("Cross-validation Results:")
print(cross_val_results.tail())
```

```python
# Analysis of the results 
 # Consistency: The small standard deviations for both train and test errors indicate that the model performs consistently across different subsets of the data.
 # Low Error Rates: The mean error rates for both train and test sets are low, suggesting that the model is making accurate predictions.
 # Minimal Overfitting: The similarity between train-error-mean and test-error-mean indicates that the model is not overfitting significantly; it performs well on both the training and validation sets.
```

```python
# Train the Final Model
```

```python
best_num_boost_round = cross_val_results.shape[0] 
final_model = xgb.train(
    params=params,
    dtrain=dtrain,
    num_boost_round=best_num_boost_round
)
```
